```python
import pandas as pd
import numpy as np
```

```python
min_sleep_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Week8_datasets/minuteSleep_merged.csv')
sleep_day_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Week8_datasets/sleepDay_merged.csv')
```

```python
# Merge the two dataset
merged_df = pd.merge(min_sleep_df, sleep_day_df, on='Id', how = 'inner')
```

```python
merged_df.head()
```

```python
merged_df.isna().sum()
```

```python
# Checking for duplicate values
merged_df.duplicated().sum()
```

```python
merged_df.shape
```

```python
merged_df.drop_duplicates(inplace=True)
```

```python
merged_df.duplicated().sum()
```

```python
merged_df.shape
```

```python
# Date format changing
merged_df['date'] = pd.to_datetime(merged_df['date']).dt.strftime('%Y-%m-%d %I:%M %p')
merged_df['date'] = merged_df['date'].str.replace('-', '/')

merged_df['SleepDay'] = pd.to_datetime(merged_df['SleepDay']).dt.strftime('%Y-%m-%d %I:%M %p')
merged_df['SleepDay'] = merged_df['SleepDay'].str.replace('-', '/')
```
