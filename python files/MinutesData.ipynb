```python
import pandas as pd
import numpy as np
```

```python
minutes_calories_narrow_df = pd.read_csv('/content/minuteCaloriesNarrow_merged.csv')
minutes_intensity_narrow_df = pd.read_csv('/content/minuteIntensitiesNarrow_merged.csv')
minutes_steps_narrow_df = pd.read_csv('/content/minuteStepsNarrow_merged.csv')
minutesMETsnarrow_df = pd.read_csv('/content/minuteMETsNarrow_merged.csv')

# Merge dataset
merged_df = pd.merge(minutes_calories_narrow_df, minutes_intensity_narrow_df, on=['Id', 'ActivityMinute'], how = 'outer')
merged_df = pd.merge(merged_df, minutes_steps_narrow_df, on=['Id', 'ActivityMinute'], how = 'outer')
minutes_activity_df = pd.merge(merged_df, minutesMETsnarrow_df, on=['Id', 'ActivityMinute'], how = 'outer')

minutes_activity_df.head()
```

```python
minutes_activity_df.drop_duplicates(inplace=True)
```

```python
minutes_activity_df.shape
```

```python
minutes_activity_df.info()
```

```python
minutes_activity_df['ActivityMinute'] = pd.to_datetime(minutes_activity_df['ActivityMinute']).dt.strftime('%Y-%m-%d %I:%M %p')
minutes_activity_df['ActivityMinute'] = minutes_activity_df['ActivityMinute'].str.replace('-', '/')
```
