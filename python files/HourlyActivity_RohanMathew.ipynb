```python
import pandas as pd
import numpy as np

```

```python
hourly_calories_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Week8_datasets/hourlyCalories_merged.csv")
hourly_intensities_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Week8_datasets/hourlyIntensities_merged.csv")
hourly_steps_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Week8_datasets/hourlySteps_merged.csv")
```

Merge the dataset with common columns

```python
merged_df = pd.merge(hourly_calories_df, hourly_intensities_df, on=['Id', 'ActivityHour'])
hourly_activity_merged_df = pd.merge(merged_df, hourly_steps_df, on=['Id', 'ActivityHour'])
```

```python
hourly_activity_merged_df.head()
```

```python
hourly_activity_merged_df['ActivityHour'] = pd.to_datetime(hourly_activity_merged_df['ActivityHour'])
```

Convert ActiveHour and format

```python
hourly_activity_merged_df['ActivityHour'] = pd.to_datetime(hourly_activity_merged_df['ActivityHour']).dt.strftime('%Y-%m-%d')
hourly_activity_merged_df['ActivityHour'] = hourly_activity_merged_df['ActivityHour'].str.replace('-', '/')
```
